```python
import re
import numpy as np
import pandas as pd
from tqdm import tqdm
```

```python
with open('big.txt','r') as fd:
    lines=fd.readlines()
    words=[]
    for line in lines:
        words+=re.findall(('\w+'),line.lower())
```

```python
words
```

### Getting Pairs

```python
def get_pairs(words):
    data=[]
    for i in range(len(words)-1):
        data.append(' '.join(words[i:i+2]))
    return data
```

```python
data = get_pairs(words)
```

```python
data
```

### Getting Probabilities

```python

prob_dist=[]
a=np.array(data)
pair,count=np.unique(a,return_counts=True)
    
print('Total Pairs: ', len(data))

# Create a list of unique pairs by converting 'pair' to a set and then back to a list
unique_pairs = list(set(data))

# Print the number of unique pairs
print('Unique Pairs: ', len(pair))

for  i in range(len(pair)):
    prob_dist.append([pair[i],count[i],pair[i].split(' ')[-1]])
print(prob_dist)
```

```python
df=pd.DataFrame(prob_dist,columns=['pair','freq','output'])
```

### Predicting

```python
df=df[df['freq']>=5]
```

```python
def predict(word):
    df_pred=[]
    for i in df.values:
        if i[0].split(' ')[0]==word:
            df_pred.append([i[0],i[1],i[2]])
    df_pred=pd.DataFrame(df_pred,columns=['input','freq','output'])
    return list(df_pred.sort_values(by='freq',ascending=False).head()['output'].values)
```

```python
predict('the')
```

### AUTO-Generation of Sequence

```python
def Auto(word):
    for i in range(10):
        pred=predict(word)
        word=pred[0]
        print(word,end=' ')
```

```python
Auto('the')
```

### Manual Sequence

```python
def Auto(word):
    preds=[]
    preds.append(word)
    for i in range(10):
        pred=predict(word)
        print(pred)
        word=pred[int(input('Enter index : '))]
        preds.append(word)
    print('-'*20)
    print(' '.join(preds))
    print('-'*20)
```
